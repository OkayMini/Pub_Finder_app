```python
import pandas as pd
import numpy as np
```

```python
pub_df = pd.read_csv('open_pubs.csv')
pub_df.head()
```

```python
pub_df.columns = ['fsa_id','name','address','postcode','easting','northing','latitude','longitude','local_authority']
```

```python
pub_df
```

```python
#Data cleaning
pub_df.isna().sum()
```

```python
# Replace "\\N" values with NaN
pub_df.replace("\\N", pd.np.nan, inplace=True)

# Convert columns to float
pub_df['latitude'] = pub_df['latitude'].astype(float)

# Convert columns to float
pub_df['longitude'] = pub_df['longitude'].astype(float)

pub_df.head()
```

```python
pub_df.isna().sum()
```

```python
pub_df = pub_df.dropna()
```

```python
pub_df.isna().sum()
```

```python
pub_df.head()
```

```python
pub_df.address
```

```python
# Extract the city name using a regular expression and create a new column 'city'
pub_df['city'] = pub_df['address'].str.extract(r',\s*([^,]+?)\s*$')
```

```python
# save the cleaned dataframe to a new CSV file
pub_df.to_csv('pub_df.csv', index=False)
```
